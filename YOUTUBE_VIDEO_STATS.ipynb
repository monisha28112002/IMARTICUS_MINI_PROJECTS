Cleaning the dataset : identifying null values, outliers etc

Remove the outliers from highly skewed columns

```python
for x in ['Trend_day_count']:
    q75,q25 = np.percentile(file.loc[:,x],[75,25])
    intr_qr = q75-q25
 
    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)
 
    file.loc[file[x] < min,x] = np.nan
    file.loc[file[x] > max,x] = np.nan
```

```python
for x in ['Tag_count']:
    q75,q25 = np.percentile(file.loc[:,x],[75,25])
    intr_qr = q75-q25
 
    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)
 
    file.loc[file[x] < min,x] = np.nan
    file.loc[file[x] > max,x] = np.nan
```
