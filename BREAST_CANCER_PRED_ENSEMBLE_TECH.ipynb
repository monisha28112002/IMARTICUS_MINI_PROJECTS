## Bagging And Boosting Technique

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df["diagnosis"]=le.fit_transform(df["diagnosis"])
df.corr().style.background_gradient(cmap="rainbow")
```

```python
import matplotlib.pyplot as plt
for i in df:
    plt.boxplot(df[i])
    plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import accuracy_score
log_classifier =  LogisticRegression()
sv_classifier = SVC()
sgd_classifier = SGDClassifier()
from sklearn.model_selection import train_test_split
y=df.iloc[:,0]
x=df.iloc[:,1:]
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=8)
```

```python
from sklearn.ensemble import VotingClassifier

vot_classifier = VotingClassifier(
    
    estimators=[('log_reg', log_classifier),
                ('svc', sv_classifier),
                ('sgd', sgd_classifier)], 
    voting='hard')

vot_classifier.fit(xtrain, ytrain)
```

```python
from sklearn.metrics import accuracy_score

def accuracy(model, data, labels):
    
    predictions = model.predict(data)
    acc = accuracy_score(labels, predictions)
    
    return acc
```

```python
accuracy(vot_classifier, xtest, ytest)
```
